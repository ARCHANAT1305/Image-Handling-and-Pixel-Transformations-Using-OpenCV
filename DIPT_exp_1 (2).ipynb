```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img_gray = cv2.imread("nature.jpg", cv2.IMREAD_GRAYSCALE)
if img_gray is None:
    raise FileNotFoundError("nature.jpg not found! Please upload it in Colab (left sidebar > Files).")

plt.imshow(img_gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")
plt.show()

# Read color image (BGR) and convert to RGB
img_bgr = cv2.imread("nature.jpg")
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title("RGB Image")
plt.axis("off")
plt.show()

# Crop part of the image
cropped = img_rgb[20:420, 200:550]
plt.imshow(cropped)
plt.title("Cropped Image")
plt.axis("off")
plt.show()

# Resize cropped image
resized = cv2.resize(cropped, (400, 400))
plt.imshow(resized)
plt.title("Resized (400x400)")
plt.axis("off")
plt.show()

# Flip horizontally
flipped_img = cv2.flip(cropped, 1)
plt.imshow(flipped_img)
plt.title("Flipped Horizontal")
plt.axis("off")
plt.show()

# Add text annotation
annotated = img_rgb.copy()
cv2.putText(annotated, "Nature Image", (50, 50),
            cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)
plt.imshow(annotated)
plt.title("Annotated Image")
plt.axis("off")
plt.show()

# Brightness adjustment
matrix = np.ones(img_rgb.shape, dtype="uint8") * 50
img_brighter = cv2.add(img_rgb, matrix)
img_darker   = cv2.subtract(img_rgb, matrix)

plt.figure(figsize=(12,4))
plt.subplot(1,3,1); plt.imshow(img_rgb); plt.title("Original"); plt.axis("off")
plt.subplot(1,3,2); plt.imshow(img_darker); plt.title("Darker"); plt.axis("off")
plt.subplot(1,3,3); plt.imshow(img_brighter); plt.title("Brighter"); plt.axis("off")
plt.show()

# Contrast adjustment
matrix1 = np.ones(img_rgb.shape, dtype="float32") * 1.1
matrix2 = np.ones(img_rgb.shape, dtype="float32") * 1.2
img_contrast1 = cv2.multiply(img_rgb.astype("float32"), matrix1)
img_contrast2 = cv2.multiply(img_rgb.astype("float32"), matrix2)
img_contrast1 = np.clip(img_contrast1, 0, 255).astype("uint8")
img_contrast2 = np.clip(img_contrast2, 0, 255).astype("uint8")

plt.figure(figsize=(12,4))
plt.subplot(1,3,1); plt.imshow(img_rgb); plt.title("Original"); plt.axis("off")
plt.subplot(1,3,2); plt.imshow(img_contrast1); plt.title("Contrast 1.1x"); plt.axis("off")
plt.subplot(1,3,3); plt.imshow(img_contrast2); plt.title("Contrast 1.2x"); plt.axis("off")
plt.show()

# Split channels (RGB)
r, g, b = cv2.split(img_rgb)
plt.figure(figsize=(12,4))
plt.subplot(1,3,1); plt.imshow(r, cmap="Reds"); plt.title("Red Channel"); plt.axis("off")
plt.subplot(1,3,2); plt.imshow(g, cmap="Greens"); plt.title("Green Channel"); plt.axis("off")
plt.subplot(1,3,3); plt.imshow(b, cmap="Blues"); plt.title("Blue Channel"); plt.axis("off")
plt.show()

# Merge channels back
merged_rgb = cv2.merge([r, g, b])
plt.figure(figsize=(8,4))
plt.subplot(1,2,1); plt.imshow(img_rgb); plt.title("Original RGB"); plt.axis("off")
plt.subplot(1,2,2); plt.imshow(merged_rgb); plt.title("Merged RGB"); plt.axis("off")
plt.show()

# HSV conversion and channels
img_hsv = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2HSV)
h, s, v = cv2.split(img_hsv)
plt.figure(figsize=(12,4))
plt.subplot(1,3,1); plt.imshow(h, cmap="hsv"); plt.title("Hue Channel"); plt.axis("off")
plt.subplot(1,3,2); plt.imshow(s, cmap="gray"); plt.title("Saturation Channel"); plt.axis("off")
plt.subplot(1,3,3); plt.imshow(v, cmap="gray"); plt.title("Value Channel"); plt.axis("off")
plt.show()

# Merge HSV back and convert to RGB
merged_hsv = cv2.merge([h, s, v])
merged_hsv_rgb = cv2.cvtColor(merged_hsv, cv2.COLOR_HSV2RGB)

plt.figure(figsize=(8,4))
plt.subplot(1,2,1); plt.imshow(img_rgb); plt.title("Original RGB"); plt.axis("off")
plt.subplot(1,2,2); plt.imshow(merged_hsv_rgb); plt.title("Merged HSVRGB"); plt.axis("off")
plt.show()
```
